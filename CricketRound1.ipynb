#*EDA AND FEATURE ENGINEERING*

```
full_df = pd.concat([train, test]).reset_index(drop = True)
```

```
target = 'Not selected'
features = full_df.columns[1:-1]
```

```
for feat in features:
    print(f'{feat}:')
    print(f'NUNIQUE: {full_df[feat].nunique()}')
    print(f'UNIQUE: {full_df[feat].unique()}')
    print('--'*40)
```

```
for col in ['Batting Avg.', 'L/R Handed Batsman']:
    full_df[col] = full_df[col].apply(lambda x: np.nan if x=='na' else x)

full_df['L/R Handed Batsman'] = full_df['L/R Handed Batsman'].apply(lambda x: 3 if x is '*' else x)

full_df['Region'] = full_df['Region'].apply(lambda x: int(x.replace('RGB', '')))
```

```
full_df.info()
```

```
for feat in features:
    print(f'{feat}:')
    print(f'{full_df[feat].value_counts()}')
    print('--'*40)
```

```
full_df['Batting Avg.'].fillna(1, inplace=True)
full_df['L/R Handed Batsman'].fillna(1, inplace=True)
```

```
from itertools import combinations

for x,y in combinations(features, 2):
    full_df[f'{x} + {y}'] = full_df[x].astype(str) + ' ' + full_df[y].astype(str)
    full_df[f'{x} + {y}'] = LabelEncoder().fit_transform(full_df[f'{x} + {y}'])
```

```
full_df[features] = full_df[features].astype(int)
```

#*TRAIN AND TEST DATA*

```
train_df = full_df[full_df[target].notna()]
test_df = full_df[full_df[target].isna()]
```

```
X = train_df.drop([target, 'player ID'], axis=1)
y = train_df[target]
```

#*MODEL*

```
def remove(features, dropped):
    remaining = [feat for feat in features if feat not in dropped]
    return remaining
```

```
skfold = StratifiedKFold(3, shuffle=True, random_state=15562)

predictions_cat = []
accuracies_cat = []
probabilities_cat = []

predictions_cat1 = []
accuracies_cat1 = []
probabilities_cat1 = []

predictions_cat2 = []
accuracies_cat2 = []
probabilities_cat2 = []

final_probabilities = []
final_accuracies = []

skfold.get_n_splits(X,y)

i = 1

for train_idx, test_idx in skfold.split(X,y):

    X_train, y_train = X.iloc[train_idx], y.iloc[train_idx]
    X_test, y_test = X.iloc[test_idx], y.iloc[test_idx]

    print(f'FOLD {i}:')

    # dropped_columns = []

    model_cat = CatBoostClassifier(iterations=1000, od_type='iter')
    model_cat.fit(X_train.drop(dropped_columns, axis=1), y_train.astype(int),
                  eval_set = (X_test.drop(dropped_columns, axis=1), y_test.astype(int)),
                  cat_features = remove(features, dropped_columns),
                  early_stopping_rounds = 100,
                  verbose = 50)
    # labels_cat = model_cat.predict(X_test)
    prob_cat = model_cat.predict_proba(X_test.drop(dropped_columns, axis=1))
    auc_cat = roc_auc_score(y_test, prob_cat[:,1])
    accuracies_cat.append(auc_cat)

    probability_cat = model_cat.predict_proba(test_df[X_train.drop(dropped_columns, axis=1).columns])[:,1]
    probabilities_cat.append(probability_cat)

    print()
    print(f'CatBoost Score: {auc_cat}')

    model_cat1 = CatBoostClassifier(iterations=3000, od_type='iter')
    model_cat1.fit(X_train.drop(dropped_columns, axis=1), y_train.astype(int),
                  eval_set = (X_test.drop(dropped_columns, axis=1), y_test.astype(int)),
                  cat_features = remove(features, dropped_columns),
                  early_stopping_rounds = 100,
                  verbose = 50)
    # labels_cat = model_cat.predict(X_test)
    prob_cat1 = model_cat1.predict_proba(X_test.drop(dropped_columns, axis=1))
    auc_cat1 = roc_auc_score(y_test, prob_cat1[:,1])
    accuracies_cat1.append(auc_cat1)

    probability_cat1 = model_cat1.predict_proba(test_df[X_train.drop(dropped_columns, axis=1).columns])[:,1]
    probabilities_cat1.append(probability_cat1)

    print()
    print(f'CatBoost Score: {auc_cat1}')

    model_cat2 = CatBoostClassifier(iterations=5000, od_type='iter')
    model_cat2.fit(X_train.drop(dropped_columns, axis=1), y_train.astype(int),
                  eval_set = (X_test.drop(dropped_columns, axis=1), y_test.astype(int)),
                  cat_features = remove(features, dropped_columns),
                  early_stopping_rounds = 100,
                  verbose = 50)
    # labels_cat = model_cat.predict(X_test)
    prob_cat2 = model_cat2.predict_proba(X_test.drop(dropped_columns, axis=1))
    auc_cat2 = roc_auc_score(y_test, prob_cat2[:,1])
    accuracies_cat2.append(auc_cat2)

    probability_cat2 = model_cat.predict_proba(test_df[X_train.drop(dropped_columns, axis=1).columns])[:,1]
    probabilities_cat2.append(probability_cat2)

    print()
    print(f'CatBoost Score: {auc_cat2}')   

    # ensemble
    prob = sum([prob_cat[:,1], prob_cat1[:,1], prob_cat2[:,1]])/3
    auc = roc_auc_score(y_test, prob)
    final_accuracies.append(auc)

    probability = sum([probability_cat, probability_cat1, probability_cat2])/3
    final_probabilities.append(probability)

    print(f'Ensemble : {auc}')
    print('----'*40)

    i += 1

print()
print(f'MAX: {max(final_accuracies)}')
print(f'MIN: {min(final_accuracies)}')
print(f'MEAN: {np.mean(final_accuracies)}')
```
